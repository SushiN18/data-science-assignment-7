```python
airline2=airline.drop(['ID#'],axis=1)
```

```python
airline.head()
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
airline_norm = norm_func(airline2.iloc[:])
```

```python
airline_norm
```

# Conducting Hierarical Clustering

```python
import scipy.cluster.hierarchy as sch
from matplotlib import pyplot as plt
```

```python
from sklearn.cluster import AgglomerativeClustering
hclusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='single')
hclusters
```

```python
AgglomerativeClustering=hclusters.fit_predict(airline_norm)
AgglomerativeClustering
```

```python
cluster_data=pd.DataFrame(AgglomerativeClustering, columns=['cluster'])
cluster_data
```

```python
airline_norm['clusters']=cluster_data
airline_norm
```

```python
from matplotlib import pyplot as plt
from sklearn .cluster import KMeans
```

```python
wcss=[]

for i in range (1,10):
    kmeans= KMeans(n_clusters=i)
    kmeans.fit(airline_norm)
    wcss.append(kmeans.inertia_)
```
